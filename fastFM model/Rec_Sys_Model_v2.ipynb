# Including recency into matrix factorization

```python
import scipy


from scipy.spatial import distance
```

```python
club_df['Days_Since_Watched'].min(),club_df['Days_Since_Watched'].max()
```

## Building the data matrix

```python
data = []


mi = club_df['Days_Since_Watched'].min()
ma = club_df['Days_Since_Watched'].max()

for u_id,h_id,d_sw in zip(club_df['UserID'],club_df['MovieID'],club_df['Days_Since_Watched']):
    scaled_d_sw = int(((d_sw-mi) /(ma-mi))*10)
    
    di = {'u_id':u_id,'h_id':h_id,'day_bin': scaled_d_sw}
    
    data.append(di)
    
#     print(u_id,h_id,scaled_d_sw)

print(len(data))
```

```python
u_ids,h_ids = {},{}

for d in data:
    u,h,da = d['u_id'],d['h_id'],d['day_bin']
    
    if not u in u_ids: u_ids[u] = len(u_ids)
    if not h in h_ids: h_ids[h] = len(h_ids)
        
        
nUsers,nHotels = len(u_ids),len(h_ids)
```

```python
nUsers,nHotels
```

```python
nDays = 10 - 0 + 1
```

```python
X = scipy.sparse.lil_matrix((len(data), nUsers + nHotels + nDays))
```

```python
for i in range(len(data)):
    user = u_ids[data[i]['u_id']]
    hotel = h_ids[data[i]['h_id']]
    days = data[i]['day_bin'] 
    
    X[i,user] = 1 # One-hot encoding of user
    X[i,nUsers + hotel] = 1 # One-hot encoding of hotels
    X[i,nUsers + nHotels + days] = 1
```

```python
y = np.array([d for d in club_df['Number_Watched_log']])

y[:10]
```

```python
X.shape,y.shape
```

```python
X.shape[0],X.shape[0]*0.7
```

```python
X_train,y_train = X[:11113],y[:11113]
X_test,y_test = X[11113:],y[11113:]
```

```python
def MSE(predictions, labels):
    differences = [(x-y)**2 for x,y in zip(predictions,labels)]
    return sum(differences) / len(differences)
```

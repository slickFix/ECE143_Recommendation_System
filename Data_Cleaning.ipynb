```python
movie_watchings = movie_watchings.filter(['WatchingID', 'ProviderID','MovieID','WatchDate'], axis=1)
movies = movies.filter(['MovieID', 'MovieType','MovieRank'], axis=1)
watchings = watchings.filter(['WatchingID', 'UserID','WatchDate','ProviderID'], axis=1)
```

Merge all these datasets into one to make it easier

WatchID - Instance of someone booking a hotel

UserID -  User ID

ProviderID - Which website/provider they used when booking (Maybe matters, maybe not?)

MovieID - HotelID

WatchDate - When the hotel was booked

MovieType - Hotel Type (By the beach, downtown, idk)

MovieRank - Hotel Stars (Not by the user, but the actual stars)

```python
df1 = pd.merge(watchings, movie_watchings, on='WatchingID')
df1 = df1.rename(columns={"ProviderID_x": "ProviderID"})
df1 = df1.rename(columns={"WatchDate_x": "WatchDate"})
df1 = df1.filter(['WatchingID', 'UserID','ProviderID','MovieID','WatchDate'], axis=1)
df2 = pd.merge(df1, movies, on='MovieID')
df2
```

```python
from datetime import datetime
# Convert the 'time' column to a datetime object
df2['WatchDate'] = pd.to_datetime(df2['WatchDate'])


# Calculate the number of days between the 'time' column and the current date
df2['Days_Since_Watched'] = (datetime.now() - df2['WatchDate']).apply(lambda x: x.days)

# Extract the month from the 'time' column
df2['Month'] = df2['WatchDate'].apply(lambda x: x.month)

df = df2
```
